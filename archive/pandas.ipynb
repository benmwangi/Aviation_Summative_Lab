```python
passengers = [
        {"name": "Alice","age": 20, "survived": 1},
        {"name": "Brett","age": 23, "survived": 0},
        {"name": "Collin","age": 25, "survived": 1}
            ]

ages = [passenger["age"] for passenger in passengers]
av = sum(ages)/len(ages)
print(av)
```

```python
import pandas as pd
data = pd.DataFrame(passengers)
av = data["age"].mean()
print(av)
```
